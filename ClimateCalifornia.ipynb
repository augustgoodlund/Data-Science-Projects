```python
import pandas as pd
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import geopandas as gpd
pd.options.display.max_rows=10
import os 
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
df=pd.read_csv("climate_data.csv")
df.head()
```

```python
df.dtypes
```

```python
#turn NaN values to 0
df['AWND'] = df['AWND'].fillna(0)
df['PRCP'] = df['PRCP'].fillna(0)
df['SNOW'] = df['SNOW'].fillna(0)
df['TAVG'] = df['TAVG'].fillna(0)
df['TMAX'] = df['TMAX'].fillna(0)
df['TMIN'] = df['TMIN'].fillna(0)
df['TSUN'] = df['TSUN'].fillna(0)
df['WDMV'] = df['WDMV'].fillna(0)
```

```python
#find averages
AWNDMean=df.groupby('DATE')['AWND'].agg(['mean'])
AWNDMean.rename(columns={'mean': 'AWNDMean'}, inplace=True)
PRCPMean=df.groupby('DATE')['PRCP'].agg(['mean'])
PRCPMean.rename(columns={'mean': 'PRCPMean'}, inplace=True)
SNOWMean=df.groupby('DATE')['SNOW'].agg(['mean'])
SNOWMean.rename(columns={'mean': 'SNOWMean'}, inplace=True)
TAVGMean=df.groupby('DATE')['TAVG'].agg(['mean'])
TAVGMean.rename(columns={'mean': 'TAVGMean'}, inplace=True)
TMAXMean=df.groupby('DATE')['TMAX'].agg(['mean'])
TMAXMean.rename(columns={'mean': 'TMAXMean'}, inplace=True)
TMINMean=df.groupby('DATE')['TMIN'].agg(['mean'])
TMINMean.rename(columns={'mean': 'TMINMean'}, inplace=True)
TSUNMean=df.groupby('DATE')['TSUN'].agg(['mean'])
TSUNMean.rename(columns={'mean': 'TSUNMean'}, inplace=True)
WDMVMean=df.groupby('DATE')['WDMV'].agg(['mean'])
WDMVMean.rename(columns={'mean': 'WDMVMean'}, inplace=True)
```

```python
#create averages table
WNDPRCP=AWNDMean.merge(PRCPMean, on="DATE")
WNDPRCPSNOW=WNDPRCP.merge(SNOWMean, on="DATE")
WNDPRCPSNOWTAVG=WNDPRCPSNOW.merge(TAVGMean, on="DATE")
WNDPRCPSNOWTAVGMAX=WNDPRCPSNOWTAVG.merge(TMAXMean, on="DATE")
WNDPRCPSNOWTAVGMAXMIN=WNDPRCPSNOWTAVGMAX.merge(TMINMean, on="DATE")
WNDPRCPSNOWTAVGMAXMINSUN=WNDPRCPSNOWTAVGMAXMIN.merge(TSUNMean, on="DATE")
AvgsTable=WNDPRCPSNOWTAVGMAXMINSUN.merge(WDMVMean, on="DATE")
AvgsTable
```
